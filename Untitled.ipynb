```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Create the mountain shape
def mountain_shape(X, Y):
    return np.exp(-0.1 * (X**2 + Y**2)) * 3  # A simple mountain-like structure

# Step 2: Create the GAM-like surface (waterfall)
def gam_surface(X, Y):
    return np.exp(-0.1 * (X**2 + Y**2)) * np.sin(5*X) * np.cos(5*Y) * 1.5  # Waterfall-like surface

# Create a grid
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

# Generate the mountain and GAM surface
Z_mountain = mountain_shape(X, Y)
Z_waterfall = gam_surface(X, Y) + Z_mountain  # The waterfall surface slightly above the mountain

# Step 3: Plot the 3D mountain and surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the mountain
ax.plot_surface(X, Y, Z_mountain, cmap='terrain', edgecolor='none', alpha=0.7)

# Plot the waterfall (GAM surface)
ax.plot_surface(X, Y, Z_waterfall, cmap='Blues', edgecolor='none', alpha=0.6)

# Step 4: Adjust the view
ax.view_init(elev=20, azim=145)  # Adjust view to give a nice perspective

# Hide axes
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Set aspect ratio for a better appearance
ax.set_box_aspect([1, 1, 0.5])

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection

# Function to create a simple mountain shape
def mountain_shape(X, Y):
    return np.exp(-0.1 * (X**2 + Y**2)) * 3  # Gaussian shape to simulate a mountain

# Function to create a GAM-like surface (waterfall)
def gam_surface(X, Y):
    return np.exp(-0.1 * (X**2 + Y**2)) * np.sin(5*X) * np.cos(5*Y) * 1.5  # Modulated sine-cosine function

# Generate a grid for plotting
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

# Generate mountain and GAM surface
Z_mountain = mountain_shape(X, Y)
Z_waterfall = gam_surface(X, Y) + Z_mountain  # The waterfall surface is slightly above the mountain

# Create a figure with 2 subplots side by side
fig = plt.figure(figsize=(16, 8))

# First subplot: 3D plot of the GAM surface over the mountain
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z_mountain, cmap='terrain', edgecolor='none', alpha=0.7)
ax1.plot_surface(X, Y, Z_waterfall, cmap='Blues', edgecolor='none', alpha=0.6)
ax1.view_init(elev=60, azim=-45)  # View angle for a good perspective
ax1.set_xticks([])  # Hide axes
ax1.set_yticks([])
ax1.set_zticks([])
ax1.set_title("3D GAM Surface and Mountain")

# Second subplot: Realistic 2D mountain drawing with rainfall and streamflow
ax2 = fig.add_subplot(122)

# Drawing the mountain with smooth curves
mountain_x = np.array([0.1, 0.3, 0.5, 0.7, 0.9])
mountain_y = np.array([0.1, 0.6, 0.9, 0.65, 0.15])
ax2.plot(mountain_x, mountain_y, color='saddlebrown', lw=3)

# Fill the mountain with color
mountain_poly = Polygon(np.column_stack([mountain_x, mountain_y]), closed=True, color='peru', alpha=0.7)
ax2.add_patch(mountain_poly)

# Adding blue patches to simulate water flow inside the mountain
water_inside = Polygon(np.array([[0.5, 0.8], [0.55, 0.6], [0.5, 0.4], [0.45, 0.6]]), closed=True, color='blue', alpha=0.5)
ax2.add_patch(water_inside)

# Adding water flowing out as a stream
ax2.plot([0.5, 0.65], [0.4, 0.2], color="blue", lw=3)  # Streamflow coming out
ax2.plot([0.65, 0.9], [0.2, 0.1], color="blue", lw=3)  # Streamflow joining the river

# Drawing rainfall as blue arrows
for i in range(4):
    ax2.arrow(0.3 + 0.1 * i, 0.9, 0, -0.15, head_width=0.02, head_length=0.03, fc='blue', ec='blue')

# Labels and annotations
ax2.text(0.3, 0.95, "Rainfall", fontsize=12, color="blue")
ax2.text(0.52, 0.55, "Infiltration", fontsize=12, color="blue")
ax2.text(0.7, 0.25, "Streamflow", fontsize=12, color="blue")

# Setting limits and hiding the axes
ax2.set_xlim(0, 1)
ax2.set_ylim(0, 1)
ax2.axis('off')  # Hide axes
ax2.set_title("Mountain with Rainfall and Streamflow")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to create a mountain surface
def mountain_surface(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) * np.exp(-0.1 * (x**2 + y**2))

# Generate data for the mountain
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)
X, Y = np.meshgrid(x, y)
Z = mountain_surface(X, Y)

# Create figure and 3D axis
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the mountain surface
ax.plot_surface(X, Y, Z, cmap='terrain', edgecolor='none', alpha=0.9)

# Add some rainfall by plotting blue dots on the mountain surface
rain_x = np.random.uniform(-5, 5, 200)
rain_y = np.random.uniform(-5, 5, 200)
rain_z = mountain_surface(rain_x, rain_y) + 0.1  # Rain falls just above the surface
ax.scatter(rain_x, rain_y, rain_z, color='blue', s=10, alpha=0.6, label='Rainfall')

# Add streamflow by drawing a line representing water flow from the top
stream_x = np.linspace(-1, 1, 100)
stream_y = stream_x
stream_z = mountain_surface(stream_x, stream_y) - 0.1
ax.plot3D(stream_x, stream_y, stream_z, color='blue', lw=3, label='Streamflow')

# Labels and title
ax.set_title('3D Mountain with Rainfall and Streamflow', fontsize=16)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Hide axes for a cleaner look
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to create a mountain surface
def mountain_surface(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) * np.exp(-0.1 * (x**2 + y**2))

# Generate data for the mountain
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)
X, Y = np.meshgrid(x, y)
Z = mountain_surface(X, Y)

# Create a mask for the lower part of the surface
angle = np.arctan2(Y, X) * 180 / np.pi  # Angle in degrees
mask = (Z < 0.03) & (angle >= -100) & (angle <= 360)

# Create figure and 3D axis
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the lower part with green color map
# Note: For simplicity, we plot the surface with a uniform color for the masked region.
# We'll manually create the green color for the masked part.
Z_green = np.where(mask, Z, np.nan)  # Use NaN to ignore regions in plotting


cmap = plt.get_cmap('terrain')

# Plot the green surface
ax.plot_surface(X, Y, Z_green, color=cmap(0.3), edgecolor='none', alpha=0.9)

# Plot the rest of the surface with the original color map
Z_rest = np.where(~mask, Z, np.nan)  # Use NaN to ignore regions in plotting

# Plot the rest of the surface
ax.plot_surface(X, Y, Z_rest, cmap='terrain', edgecolor='none', alpha=0.9)

# Add some rainfall by plotting blue dots on the mountain surface
rain_x = np.random.uniform(-5, 5, 300)
rain_y = np.random.uniform(-5, 5, 300)
rain_z = mountain_surface(rain_x, rain_y) + 0.3  # Rain falls just above the surface
ax.scatter(rain_x, rain_y, rain_z, color='blue', s=10, alpha=0.6, label='Rainfall')

# Add streamflow by drawing a line representing water flow from the top
stream_x = np.zeros(100)
stream_y = np.linspace(-5, -1, 100)
stream_z = mountain_surface(stream_x, stream_y) +0.2
#ax.plot3D(stream_x, stream_y, stream_z, color='blue',  lw=10, label='Streamflow')

# Labels and title
ax.set_title('3D Mountain with Rainfall and Streamflow', fontsize=16)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Hide axes for a cleaner look
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_axis_off()
ax.w_xaxis.pane.fill = False
ax.w_yaxis.pane.fill = False
ax.w_zaxis.pane.fill = False
ax.w_xaxis.line.set_color((1.0, 1.0, 1.0, 0.0))
ax.w_yaxis.line.set_color((1.0, 1.0, 1.0, 0.0))
ax.w_zaxis.line.set_color((1.0, 1.0, 1.0, 0.0))


# Add a legend
ax.legend(loc='best', bbox_to_anchor=(0.8, 0.8))
ax.view_init(elev=25, azim=185)  # Adjust view to give a nice perspective


# Show the plot
plt.show()
```
