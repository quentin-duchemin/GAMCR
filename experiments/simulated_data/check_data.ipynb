## Preprocessing simulated data

```python
import pandas as pd
import numpy as np
station = 'Lugano'
mode = 'flashy'
data_path = '/mydata/watres/quentin/code/FLOW/data/simulated_data_events_updatesummer2024/{0}_{1}/ref_flow_{1}_fluxes.txt'.format(station, mode)
#'./{0}_{1}/data_{0}_{1}.txt'.format(station, mode)
df = pd.read_csv(data_path)
df
```

```python
import sys
sys.path.append('/mydata/watres/quentin/code/FLOW/hourly_analysis/')
import GAMCR
```

```python
df = df.rename(columns={"t": "timeyear"})
df['date'] = df['timeyear'].apply(lambda x: GAMCR.fractional_year_to_datetime(x))
import os
if not os.path.exists('./{0}_{1}/'.format(station, mode)):
    os.makedirs('./{0}_{1}/'.format(station, mode))
data_path_save = './{0}_{1}/data_{0}_{1}.txt'.format(station, mode)
df.to_csv(data_path_save, index=False)
```
